```python
def letter_substitution(data):
    
    data = data.rename(columns={data.columns[0]: 'Артикул'})
    data['Артикул'] = data['Артикул'].astype(str)
    data = data.reset_index(drop=True)
    
    fa = list(filter(lambda x: re.compile('[^а-яА-Я ]').sub('', x), list(data['Артикул'])))

    if len(fa) > 0:

        Q = {'е': 'e', 
             'Е': 'E', 
             'Т': 'T', 
             'у': 'y', 
             'о': 'o', 
             'О': 'O', 
             'р': 'p', 
             'Р': 'P', 
             'а': 'a', 
             'А': 'A', 
             'Н': 'H',
             'К': 'K',
             'х': 'x',
             'Х': 'X',
             'с': 'c',
             'С': 'C',
             'В': 'B',
             'М': 'M'}

        for i in range(len(data)):

            for row in list(Q.keys()):

                if row in data.loc[i]['Артикул']:

                    data['Артикул'].loc[i] = data.loc[i]['Артикул'].replace(row, Q[row])


    fa2 = list(filter(lambda x: re.compile('[^а-яА-Я ]').sub('', x), list(data['Артикул'])))

    if len(fa2) == 0:

        return data




```

```python
# Расчет итоговой наценки
def itogo_nac(data):
    data2 = data.copy()
    if 'Цена' not in data2.columns:
        print('Нет цены')
        exit()
    if 'Наценка' not in data2.columns:
        print('Нет наценки')
        exit()
    if 'Выручка' not in data2.columns:
        data2['Выручка'] = data2[[y.strftime('%d-%B-%Y') \
                                for y in sorted([pd.to_datetime(x) for x in dt_col(data2)])][-1]] * data2['Цена']
    data2 = data2[(data2['Наценка'].astype(str).isin(['inf', '-inf', 'nan', 'None'])==False) & \
              (data2['Выручка'].astype(str).isin(['nan', 'None'])==False)][['Наценка', 'Выручка']]
    if sum(data2['Выручка']) > 0:
        data2['Доля выручки'] = (data2['Выручка'] * 100) / sum(data2['Выручка'])
        data2['Итого наценка'] = data2['Доля выручки'] / 100 * data2['Наценка']
        S = sum(data2['Итого наценка'])
    else:
        S = 0 
    return S
```

```python
def create_metrcs(path_metrics,  
                  metrics_name=metrics_name, 
                  flag=0, 
                  COLS_DEL=COLS_DEL,
                  Col_for_not_del=Col_for_not_del,
                  flag_marketing=False):
    
    color_dict = {'красный': 'ffafb7',
                  'зеленый': 'a6faa7',
                  'голубой': 'bff2ff',
                  'желтый': 'fafc97',
                  'фиолетовый': 'c8aef8'}
    
    template = openpyxl.load_workbook(path_metrics)
    
    for i in list(range(len(template.sheetnames[:-1]))):
        sheet_name = template.sheetnames[i]
        ws = template.get_sheet_by_name(sheet_name)
        # Устанавливаем фильтры на столбцы
        ws.auto_filter.ref = ws.dimensions
        # Ограничиваем ширину поля с именем товара
        if sheet_name in ['WB', 'ozon', 'yandex', 'sber', 'WB_доп', 'ozon_доп', 'ozon_ssy', 'ozon_kz', 'wb_kz']:
            ws.column_dimensions['B'].width = 30
        elif sheet_name == 'letual':
            ws.column_dimensions['C'].width = 30
        elif sheet_name == 'ozon2':
            ws.column_dimensions['B'].width = 30
        else:
            None

        df=pd.read_excel(path_metrics, sheet_name=sheet_name)
        
        
        # Первый блок изменений - базовый
        for col in df.columns:
            if col == 'Выручка':
                df['Выручка'] = df['Выручка'].round(0)
            elif col == 'Баркод':
                ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].hidden= True
            elif col == 'ТМ':
                ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].hidden= True
            #Заливка статусов
            elif col == 'Статус':
                for j in range(1, df.shape[0]+1):
                    cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=j)
                    if str(cel.value) == 'отказ' or str(cel.value) == 'отказ 2':
                        cel.font = Font(color = "9A5400")
                        cel.fill = PatternFill('solid', fgColor = color_dict['желтый'])
                    if str(cel.value) == 'архив':
                        cel.font = Font(color = "A91B1F")
                        cel.fill = PatternFill('solid', fgColor = color_dict['красный'])
                    if 'новинка' in str(cel.value):
                        cel.font = Font(color = "196605")
                        cel.fill = PatternFill('solid', fgColor = color_dict['зеленый'])
            #Заливка колонки остатков
            elif col == 'Остаток':
                for j in range(1, df.shape[0]+1):
                    cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=j)
                    cel.fill = PatternFill('solid', fgColor = color_dict['голубой'])
            #Заливка колонок с неделями
            elif col.count('-') == 3:
                for j in range(1, df.shape[0]+1):
                    cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=j)
                    cel.fill = PatternFill('solid', fgColor = color_dict['голубой'])
            #Заливка колонки с наценками
            elif col == 'Наценка':
                for j in range(1, df.shape[0]+1):
                    cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=j)
                    if type(cel.value) in [float, int] and cel.value < 100:
                        cel.font = Font(color = "A91B1F")
                        cel.fill = PatternFill('solid', fgColor = color_dict['красный'])  
                    if type(cel.value) in [float, int] and cel.value >= 100:
                        cel.font = Font(color = "196605")
                        cel.fill = PatternFill('solid', fgColor = color_dict['зеленый'])
            #Заливка колонки с наценками
            elif col == col_nac or col == col_nac2:
                for j in range(1, df.shape[0]+1):
                    cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=j)
                    if type(cel.value) in [float, int] and cel.value < 100:
                        cel.font = Font(color = "A91B1F")
                        cel.fill = PatternFill('solid', fgColor = color_dict['красный'])  
                    if type(cel.value) in [float, int] and cel.value >= 100:
                        cel.font = Font(color = "196605")
                        cel.fill = PatternFill('solid', fgColor = color_dict['зеленый'])
            #Заливка артикулов
            elif col == 'Артикул':
                #На вб
                if sheet_name in ['WB', 'WB_доп']:
                    for j in range(1, df.shape[0]+1):
                        cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=j)
                        if cel.value in spr_wb_blue:
                            cel.fill = PatternFill('solid', fgColor = color_dict['голубой'])
                        elif cel.value in spr_wb_green:
                            cel.fill = PatternFill('solid', fgColor = color_dict['зеленый'])
                        elif cel.value in spr_wb_violet:
                            cel.fill = PatternFill('solid', fgColor = color_dict['фиолетовый'])
                        elif cel.value in spr_wb_red:
                            cel.fill = PatternFill('solid', fgColor = color_dict['красный'])
                        else:
                            None
                #На озоне
                elif sheet_name in ['ozon', 'ozon_доп']:
                    for j in range(1, df.shape[0]+1):
                        cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=j)
                        if cel.value in spr_oz_green:
                            cel.fill = PatternFill('solid', fgColor = color_dict['зеленый'])
                        elif cel.value in spr_oz_violet:
                            cel.fill = PatternFill('solid', fgColor = color_dict['фиолетовый'])
                        elif cel.value in spr_oz_red:
                            cel.fill = PatternFill('solid', fgColor = color_dict['красный'])
                        else:
                            None
                #На яндексе
                elif sheet_name == 'yandex':
                    for j in range(1, df.shape[0]+1):
                        cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=j)
                        if cel.value in ya_art_green:
                            cel.fill = PatternFill('solid', fgColor = color_dict['зеленый'])
                        else:
                            None
                    
            #Статус рекламы, заливка
            elif col == 'статус рекламы':
                #На вб
                if sheet_name in ['WB', 'WB_доп']:
                    for j in range(1, df.shape[0]+1):
                        cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=j)
                        if cel.value == 'активна':
                            cel.fill = PatternFill('solid', fgColor = color_dict['зеленый'])
                        elif cel.value == 'приостановлена':
                            cel.fill = PatternFill('solid', fgColor = color_dict['красный'])
                        else:
                            None  
            #Наименование - название, заливка
            elif col == 'Наименование' or col == 'Название':
                #На вб и озон
                if sheet_name in ['WB', 'WB_доп']:
                    for j in range(1, df.shape[0]+1):
                        column_with_articul_number = list(df.columns).index('Артикул')+1
                        cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=j)
                        cel_art = template[sheet_name].cell(column=column_with_articul_number, row=j)
                        if cel_art.value in green_list_name_wb:
                            cel.fill = PatternFill('solid', fgColor = color_dict['зеленый'])
                        elif cel_art.value in yellow_list_name_wb:
                            cel.fill = PatternFill('solid', fgColor = color_dict['желтый'])
                        elif cel_art.value in blue_list_name_wb:
                            cel.fill = PatternFill('solid', fgColor = color_dict['голубой'])
                        else:
                            None  
                elif sheet_name in ['ozon', 'ozon_доп']:
                    for j in range(1, df.shape[0]+1):
                        column_with_articul_number = list(df.columns).index('Артикул')+1
                        cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=j)
                        cel_art = template[sheet_name].cell(column=column_with_articul_number, row=j)
                        if cel_art.value in green_list_name_oz:
                            cel.fill = PatternFill('solid', fgColor = color_dict['зеленый'])
                        elif cel_art.value in yellow_list_name_oz:
                            cel.fill = PatternFill('solid', fgColor = color_dict['желтый'])
                        elif cel_art.value in blue_list_name_oz:
                            cel.fill = PatternFill('solid', fgColor = color_dict['голубой'])
                        else:
                            None  
            #Оценка IDC, заливка
            elif col == 'Оценка IDC':
                #На озоне
                if sheet_name in ['ozon', 'ozon_доп']:
                    for j in range(1, df.shape[0]+1):
                        cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=j)
                        if cel.value == 'Критическая':
                            cel.fill = PatternFill('solid', fgColor = color_dict['красный']) 
            elif col == 'Номенклатура' or col == 'Номенклaтурa':
                #На вб
                if sheet_name in ['WB', 'WB_доп']:
                    for j in range(1, df.shape[0]+1):
                        cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=j)
                        if str(cel.value) not in ['None', 'nan', 'Nan', 'NaN']:
                            cel.hyperlink = "https://www.wildberries.ru/catalog/" + str(cel.value) + "/detail.aspx"
                #На озоне
                elif sheet_name in ['ozon', 'ozon_доп']:
                    for j in range(1, df.shape[0]+1):
                        cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=j)
                        if str(cel.value) not in ['None', 'nan', 'Nan', 'NaN']:
                            cel.hyperlink = "https://www.ozon.ru/product/" + str(cel.value)
                
                
        # Второй блок изменений (метрики для телеграмма и "Цены") - редакция Даши
        if flag == 1:
            #Вычленяем Общую итоговую наценку
            itogo_nac = 0
            if 'Итого наценка' in df.columns and sheet_name in \
            ['WB', 'ozon', 'yandex', 'sber', 'letual', 'ozon_ssy', 'ozon_kz', 'wb_kz']:
                column_number = list(df.columns).index('Итого наценка')+1
                cel = template[sheet_name].cell(column=column_number, row=df.shape[0]+1)
#                 cel.fill = PatternFill('solid', fgColor = color_dict['желтый']) 
                itogo_nac = round(cel.value, 2)
                print(sheet_name, itogo_nac)
            if 'full_df_3.0' in path_metrics:
                for row in COLS_DEL:
                    if row not in Col_for_not_del:
                        if row in df.columns:
                            ws.delete_cols(list(df.columns).index(row)+1, 1)
                            df = df.drop(row, axis=1)
            else:
                for row in COLS_DEL:
                    if row in df.columns:
                        ws.delete_cols(list(df.columns).index(row)+1, 1)
                        df = df.drop(row, axis=1)
            if sheet_name in ['WB', 'ozon']:
                ws = template.get_sheet_by_name(sheet_name)
                ws.freeze_panes = ws['H2']
            elif sheet_name in ['yandex']:
                ws = template.get_sheet_by_name(sheet_name)
                ws.freeze_panes = ws['E2']
            elif sheet_name in ['sber', 'ozon_ssy', 'ozon_kz', 'wb_kz']:
                ws = template.get_sheet_by_name(sheet_name)
                ws.freeze_panes = ws['F2']
            elif sheet_name in ['letual', 'WB_доп', 'ozon_доп']:
                ws = template.get_sheet_by_name(sheet_name)
                ws.freeze_panes = ws['G2']
            else:
                None  
            ws.auto_filter.ref = ws.dimensions
            if sheet_name in ['WB', 'ozon', 'yandex', 'sber', 'letual', 'ozon_ssy', 'ozon_kz', 'wb_kz']:
                for col in df.columns:
                    #Последнюю строку ("Итого") в датафрейме заменяем на "None"
                    cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=df.shape[0]+1)
                    cel.value = None
                    if col == 'Наценка':
                        if 'full_df_3.0' not in path_metrics:
                            cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=1)
                            if itogo_nac != 0:
                                cel.value = str(itogo_nac) # 'размер средневзвешанной итоговой наценки'
                        ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 12
                    elif col == 'Остаток на РЦ':
                        cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=1)
                        cel.value = 'Ост.Химки'
                        ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 16
                    elif col == 'Остаток РЦ Иркутск':
                        cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=1)
                        cel.value = 'Ост.Иркутск'
                        ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 16
                    elif col == 'В резерве':
                        cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=1)
                        cel.value = 'Резерв'
                        ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 12
                    elif col == 'Спринт вб':
                        cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=1)
                        cel.value = 'раздачи ВБ'
                        ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 16
                    elif col == 'Спринт озон':
                        cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=1)
                        cel.value = 'раздачи Озон'
                        ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 16
                    elif col == 'Склад_Иркутск' or col == 'Склад Иркутск':
                        cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=1)
                        cel.value = 'FBS Иркутск'
                        ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 16
                    elif col == 'Склад_Поставщик_Химки' or col == 'Склад Химки (FBS обычные товары)':
                        cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=1)
                        cel.value = 'FBS Химки'
                        ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 14
                    elif col == 'ПСР':
                        cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=1)
                        cel.value = 'FBS ПСР'
                        ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 12
                    elif col == 'прибыль с единицы товара':
                        cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=1)
                        ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 14
                        for j in range(1, df.shape[0]+1):
                            cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=j)
                            cel.number_format = '0'
                    elif col == 'Статус':
                        ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 16
                    elif col == 'ТМ':
                        ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 16
                    elif col == 'Баркод':
                        ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 16
                    elif col == 'Moscow_warehouses':
                        ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 16
                    elif col == 'Остаток в Невинномысске':
                        cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=1)
                        cel.value = 'Ост.Невинномысск'
                        ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 12
                    elif col == 'Закончится':
                        ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 14
                    elif col == 'ждём, шт.':
                        ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 12
                    elif col == 'Маркетинг':
                        ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 8
                    elif col == 'Выручка':
                        ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 12
                    elif col == 'IDC':
                        ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 8
                    elif col == 'Номенклатура':
                        ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 11
                    elif col == 'Номенклaтурa':
                        ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 11
                    elif col == 'Себестоимость':
                        ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 11
                    elif col == 'Артикул':
                        ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 6 
                        if 'full_df_3.0' not in path_metrics:
                            for j in range(1, df.shape[0]+1):
                                cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=j)
                                try:
                                    cel.value = int(cel.value)
                                except:
                                    None
                                cel.number_format = '0'
                    try:
                        pd.to_datetime(col)
                        cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=1)
                        if 'full_df_3.0' not in path_metrics:
                            cel.value = pd.to_datetime(col)
                            cel.number_format = 'd.m'
                        else:
                            cel.value = pd.to_datetime(col).strftime('%d.%m')
                        cel.alignment = Alignment(horizontal='left')
                        ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 8
                    except:
                        if '-' in col:
                            try:
                                pd.to_datetime('-'.join(col.split('-')[:2])+'-2023')
                                cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=1)
                                cel.value = pd.to_datetime('-'.join(col.split('-')[:2])+'-2023')\
                                            .strftime('%d.%m') + \
                                            '-' + pd.to_datetime('-'.join(col.split('-')[2:])+'-2023')\
                                            .strftime('%d.%m')
                                if 'full_df_3.0' not in path_metrics:
                                    cel.number_format = 'd.m' + '-' + 'd.m'
                                cel.alignment = Alignment(horizontal='left')
                                ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 12
                            except:
                                None   
                                

            # Третий блок изменений (для "Цен") - редакция Алены
            if 'full_df_3.0' in path_metrics:
                if sheet_name in ['WB', 'ozon', 'yandex', 'sber', 'ozon_kz', 'wb_kz']:
                    for col in df.columns:
                        if col in ['Артикул', 'Наименование', 'Статус']:
                            for j in range(1, df.shape[0]+1):
                                cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=j)
                                cel.number_format = ''
                        if col in Col_for_not_del and 'Цена' in col or 'Наценка' in col:
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 18
                        elif 'Индекс цен' in col:
                            for j in range(1, df.shape[0]+1):
                                cel = template[sheet_name].cell(column=list(df.columns).index(col)+1, row=j)
                                if cel.value == 'Невыгодный':
                                    cel.font = Font(color = "A91B1F")
                                    cel.fill = PatternFill('solid', fgColor = color_dict['красный'])
                                elif cel.value == 'Умеренный':
                                    cel.font = Font(color = "9A5400")
                                    cel.fill = PatternFill('solid', fgColor = color_dict['желтый'])
                                else:
                                    None 
                        elif col == 'В резерве':
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].hidden= True
                        elif col == 'Склад_Иркутск' or col == 'Склад Иркутск':
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].hidden= True
                        elif col == 'Склад_Поставщик_Химки' or col == 'Склад Химки (FBS обычные товары)':
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].hidden= True
                        elif col == 'ПСР':
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].hidden= True
                        elif col == 'Выручка':
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].hidden= True
                        elif col == 'Moscow_warehouses':
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].hidden= True
                        elif col == 'IDC':
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].hidden= True
                        elif col == 'прибыль с единицы товара':
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].hidden= True
                        elif col == 'Себестоимость':
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].hidden= True
                        elif col == 'Остаток на РЦ' and sheet_name in ['WB', 'ozon']:
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].hidden= True
                        elif col == 'Остаток РЦ Иркутск' and sheet_name in ['WB', 'ozon']:
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].hidden= True
                        elif col == 'Маркетинг':
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].hidden= True
                        elif col == 'Ценовой индекс товара на Ozon':
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].hidden= True
                        elif col == 'Ценовой индекс товара на рынке':
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].hidden= True
                        elif col == 'Ценовой индекс товара на рынке на мои товары':
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].hidden= True
                        elif col == 'Рыночная цена конкурентов, руб.':
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].hidden= True
                        elif col == 'Итого наценка__совокупная':
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 24
                        elif col == 'Установленный вами порог, ₽':
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 20
                        elif col == 'Оптимальный порог по расчётам Маркета, ₽':
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 20
                        elif col == 'Порог для привлекательной цены, ₽':
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 20
                        elif col == 'Порог для умеренно привлекательной цены, ₽':
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 20 
                        elif col == 'Ваша цена, ₽ FBY':
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 20
                        elif col == 'Ваша цена, ₽ FBS Х':
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 20
                        elif col == 'Ваша цена, ₽ FBS И':
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 20 
                        elif col == 'Статус':
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 8
                        elif col == 'ждём, шт.':
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 12
                        elif col == 'Ссылка на рыночную цену конкурентов':
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 12
                        elif col == 'Остаток' and  sheet_name in ['WB', 'ozon', 'yandex', 'sber', 'ozon_kz', 'wb_kz']:
                            cel = template[sheet_name].cell(column=list(df.columns).index(col), row=1)
                            if itogo_nac != 0:
                                cel.value = str(itogo_nac) # 'размер средневзвешанной итоговой наценки'                        
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col))].width = 12
                        elif col in ['Цена до скидки', 'Спринт вб', 'Спринт озон']:
                            ws.delete_cols(list(df.columns).index(col)+1, 1)
                            df = df.drop(col, axis=1)
                            
                    for col in df.columns:
                        if col in wbpd_dop_col:
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 8
                        elif col in [col_nac, col_nac2, 'Наценка', 'Цена', 'Оценка IDC', 'Моя цена на других площадках, руб.']:
                            ws.column_dimensions[get_column_letter(list(df.columns).index(col)+1)].width = 8
                    

                    
    if flag_marketing == True:
        for row in template.sheetnames[2:]:
            if row not in ['WB', 'ozon']:
                template.remove(template[row])
        if 'full_df_V' in path_metrics:
            template.save(path_algoritm + f"Все файлы для ежедневной презентации/Метрики_маркетинг_Вадим.xlsx")
        elif 'full_df_E' in path_metrics:
            template.save(path_algoritm + f"Все файлы для ежедневной презентации/Метрики_маркетинг_Евгений.xlsx")
                    
    # Сохранение
    if flag == 1:
        if 'full_df_2.0' in path_metrics:
            template.save(path_algoritm + f'Все файлы для ежедневной презентации/Метрики для телеграм/{metrics_name}.xlsx')
        elif 'full_df_3.0' in path_metrics:
            for row in template.sheetnames[4:]:
                if row not in ['WB', 'ozon', 'yandex', 'sber', 'ozon_kz', 'wb_kz']:
                    template.remove(template[row])
            template.save(path_algoritm + f"Все файлы для ежедневной презентации/Цены_{datetime.datetime.now().strftime(format='%d.%m')}.xlsx")
            
    else:
        if 'full_df_2.0' in path_metrics:
            if metrics_name == 'Метрики':
                template.save(path_downloads + 'Дима/' + f'{metrics_name}.xlsx')
            else:
                template.save(path_algoritm + f'Все файлы для ежедневной презентации/{metrics_name}.xlsx')
        else:
            for row in template.sheetnames[4:]:
                template.remove(template[row])
            template.save(path_algoritm + f"Все файлы для ежедневной презентации/Цены_{datetime.datetime.now().strftime(format='%d.%m')}.xlsx")
                    

                            
                            
                            
                            
                            
                            
                            
                            
```
